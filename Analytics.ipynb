```
import pandas as pd
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
%matplotlib inline
from google.colab import files
uploaded = files.upload()
```

```
df = pd.read_csv("Data1.csv")#enter the name of the file
df=df.head(37)#enter the number of rows (zero index)
df
```

```
Questions=['Q1','Q2','Q3','Q4','Q5','Q6','Q7','Q8','Q9 (Y/N)']# Name of the Questions
for i in Questions:
  print()
  print("Quention Number",i)
  list1=df[i].values.tolist()
  print("Responces=",list1)
  new_list = [x for x in list1 if str(x) != 'nan']
  lists = sorted(Counter(new_list).items())
  length_of_list=len(new_list)
  print("Number of people who answered this question=",length_of_list)
  x, y = zip(*lists)
  #print(x,y)
  Dict = {1: 'Strongly_Agree', 2: 'Agree', 3: 'Neutral',4: 'Disagree',5:'Strongly_Disagree'} 
  z=[]
  for i in x:
    z.append(Dict[int(i)])
  plt.bar(z, y)
  plt.show()
  for i,j in lists:
    if i == 1.0:
      print(int((j/length_of_list)*100),"Percent of the people Strongly agree with this question",'(',j,')')
    elif i==2.0:
      print(int((j/length_of_list)*100),"Percent of the people Agree with this question",'(',j,')')
    elif i==3.0:
       print(int((j/length_of_list)*100),"Percent of the people have a neutral opinion",'(',j,')')
    elif i==4.0:
       print(int((j/length_of_list)*100),"Percent of the people Disgree with this question",'(',j,')')
    else:
      print(int((j/length_of_list)*100),"Percent of the people Strongly Disagree with this question",'(',j,')')

```
